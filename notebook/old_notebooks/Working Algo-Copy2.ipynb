```python
# Loading up libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading up csv
df = pd.read_csv(r"./csv/data_3600.csv")
```

```python
SMAL = df['Price'].rolling(window = 5).mean()
SMAH = df['Price'].rolling(window = 25).mean()
```

```python
df["SMAL"] = SMAL
df["SMAH"] = SMAH
df["Change"] = ""
df["Buy/Sell"] = ""
df["TimeLimit"] = 0
df["isAllowed"] = False
df["Action"] = "---"
```

```python
def checkCrossover(row):
    if row["SMAL"] > row["SMAH"]:
        return "Higher"
    else:
        return "Lower"
def isBuyOrSell(row):
    if row["Crossover"] == "Higher":
        return "Buy"
    else:
        return "Sell"
    
def tradingAction(row):
    if row["isAllowed"] and row["Change"] == "Crossed":
        if row["Buy/Sell"] == "Buy":
            return "Buy!"
        else:
            return "Sell!"
```

```python
df["Crossover"] = df.apply(checkCrossover, axis=1
```

```python
for index, row in df.iterrows():
    if index == 0:
        df["Change"][index] = "N/A"
    else:
        if df["Crossover"][index] == df["Crossover"][index-1]:
            df["Change"][index] = "---"
        else:
            df["Change"][index] = "Crossed"

for index, row in df.iterrows():
    if index == 0:
        df["Buy/Sell"][index] = True
    else:
        __currentMode = df["Buy/Sell"][index - 1]
        df["Buy/Sell"][index] = __currentMode
        if df["Change"][index] == "Crossed":
            df["Buy/Sell"][index] = not __currentMode


            
# for index, row in df.iterrows():
#     if index == 0:
#         df["TimeLimit"][index] = 0
#     else:
        
```

```python
for index, row in df.iterrows():
    if index == 0:
        df["TimeLimit"][index] = 0
    else:
        __TimeLimit = df["TimeLimit"][index - 1]
        __isAllowed = df["isAllowed"][index - 1]
        df["TimeLimit"][index] = __TimeLimit + 1
        if df["TimeLimit"][index] > 60:
            df["TimeLimit"][index] = 0
        if df["TimeLimit"][index] >= 30 and df["TimeLimit"][index] <= 60:
            df["isAllowed"][index] = True
        else:
            df["isAllowed"][index] = False


for index, row in df.iterrows():
    if index == 0:
        df["Action"][index] = "Buy!"
    if df["isAllowed"][index] and df["Change"][index] == "Crossed":
        if df["Buy/Sell"][index]:
            df["Action"][index] = "Buy!"
        else:
            df["Action"][index] = "Sell!"
        df["TimeLimit"][index] = 0


        
```
